```python
# time asd amount are not proper so we are dooing scaling: we are using robust scalar
```

```python
from sklearn.preprocessing import RobustScaler
scalar=RobustScaler()
data["Time"]=scalar.fit_transform(data[["Time"]])
data["Amount"]=scalar.fit_transform(data[["Amount"]])
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
#now  two ways, we an do undersampling or oversampling . so now iam doing oversampling
```

```python
x=data.drop("Class",axis=1)
y=data["Class"]
```

```python
y.shape
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=42)
x,y=ros.fit_resample(x,y)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
def model(model):
  model.fit(x_train,y_train)
  pred=model.predict(x_test)
  from sklearn.metrics import accuracy_score
  print(accuracy_score(y_test,pred)*100)
```

```python
from sklearn.linear_model import LogisticRegression
model(LogisticRegression())
```

```python
from sklearn.tree import DecisionTreeClassifier
model(DecisionTreeClassifier())
```

```python
from sklearn.ensemble import RandomForestClassifier
model(RandomForestClassifier())
```

```python
from xgboost import XGBClassifier
model(XGBClassifier())
```

```python
#XGBClassifer has best accuracy
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Use f_classif for data with negative values
select = SelectKBest(f_classif, k=5)
select.fit(x_train, y_train)
x_train.columns[select.get_support()]
```
